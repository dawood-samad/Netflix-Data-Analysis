```python
import pandas as pd
df = pd.read_csv('netflix_titles.csv')
df.head()
```

```python
print(df.shape)
```

```python
# Count top 10 most frequent directors
df['director'].value_counts().head(10)
```

```python
# Remove extra spaces from 'date_added' and convert to datetime
df['date_added'] = pd.to_datetime(df['date_added'].str.strip(), format="%B %d, %Y", errors='coerce')

# Now extract the year
df['year_added'] = df['date_added'].dt.year

# Count how many titles were added each year
df['year_added'].value_counts().sort_index().plot(kind='bar', title='Titles Added Each Year', figsize=(10,5))
```

```python
# Count how many titles were released each year
df['release_year'].value_counts().sort_index().plot(kind='line', title='Content Release Trend', figsize=(10,5))
```

```python
# Add a new column for title length
df['title_length'] = df['title'].apply(len)

# Plot distribution of title lengths
df['title_length'].plot(kind='hist', bins=20, title='Distribution of Title Lengths', figsize=(10,5))
```
